```python
import pandas as pd
import matplotlib.pyplot as plt

# Excel 파일 로드
distance_file = 'direct_node_distances.xlsx'
throughput_file = 'rxsite_throughput_with_node_id.xlsx'

# pandas를 사용하여 파일 읽기
distance_df = pd.read_excel(distance_file)
throughput_df = pd.read_excel(throughput_file)

# 거리와 처리율의 통계 값 계산
distance_stats = distance_df['Distance (m)'].describe()
throughput_stats = throughput_df['Throughput(Mbps)'].describe()

# 거리와 처리율의 분포 시각화
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# 거리 분포 시각화
axes[0].hist(distance_df['Distance (m)'], bins=20, color='skyblue', edgecolor='black')
axes[0].set_title('Distance Distribution')
axes[0].set_xlabel('Distance (m)')
axes[0].set_ylabel('Frequency')

# 처리율 분포 시각화
axes[1].hist(throughput_df['Throughput(Mbps)'], bins=20, color='lightgreen', edgecolor='black')
axes[1].set_title('Throughput Distribution (Mbps)')
axes[1].set_xlabel('Throughput (Mbps)')
axes[1].set_ylabel('Frequency')

# 시각화 출력
plt.tight_layout()
plt.show()

# 계산된 통계값 출력
print("거리 통계:")
print(distance_stats)
print("\n처리율 통계:")
print(throughput_stats)
```
